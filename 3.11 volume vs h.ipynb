### 3.11 원통형 저장고 내부액체의 부피

#### $$ V = [r^2\cos^{-1}(\frac{r - h}{r}) - (r - h)\sqrt{2rh - h^2}]L $$

* V = 저장고 내부의 액체의 부피
* r = 반지름
* L = 길이
* h = 원통 중심선으로 부터 액체의 깊이

```python
def cyltank(r,L):
    """
    실린더의 부피를 구하는 함수
     
    paramerters
    -----------
    r -> int or float
    실린더의 반지름  
    L -> 
    plot_title = 
    return
    h = 
    ------ -----
    V 
    """    
    global h
    h = np.linspace(0,6,50)     
    return (r**2*np.arccos((r - h)/r) - (r -h)*np.sqrt(2*r*h - h**2)) * L
    
```

```python
cyltank(3,5)
```

```python
plt.plot(h,cyltank(3,5),c = 'r')
plt.title =  ("Volume vs. Depth for Horizontal cylindrical Tank")
plt.xlabel = ('h')
plt.ylabel = ('V')
plt.show()
```

### 3.17 윤년 계산하기

```python
ly = []
def leap_year(a,b):
    year = np.array(np.arange(a,b))
    for i in year:
        if i % 4 == 0:
            ly.append(i)            
        
        elif i % 4 == 0 and i % 400:
            pass
        else:
            pass
    print(f'{a}년과 {b}년 사이의 윤년의 개수는 {len(ly)}개 입니다')
leap_year(0,2023)
```

### 3.25 transpose temperature

$$ \frac{(F - 32)}{1.8} = C $$
$$ (C \times 1.8) + 32 = F $$

* F = 화씨온도
* C = 섭씨온도

```python
import warnings
warnings.filterwarnings('ignore')

day = np.array([15,45,75,105,135,165,195,225,255,285,315,345])
tfdv = np.array([54,60,69,77,87,96,102,101,92,78,63,52])
tcsp = np.array([-27,-40,-53,-56,-57,-57,-59,-59,-59,-50,-38,-27])
n = len(day)
tcdv = np.zeros(n)
tfsp = np.zeros(n)
def tempconv(temperature):    
    if temperature in tfdv:
        for i in range(len(day)):        
            C = (temperature[i] - 32)/1.8
            tcdv[i] = C
                         
    
    elif temperature in tcsp:
        for i in range(len(day)):
            F = (temperature[i] * 1.8) + 32
            tfsp[i] = F    
        
    
    else:
        print('Flase')
```

```python
tempconv(tfdv)
tempconv(tcsp)
fig, (ax1,ax2) = plt.subplots(figsize = (10,6), ncols = 2, nrows = 1)
ax1.scatter(tfdv,tcdv)
ax2.scatter(tcsp,tfsp)
```
