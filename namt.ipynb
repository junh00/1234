### 구간법

#### 이분법
* $ f(x) $ 의 부호가 바뀌는 구간을 찾는 방법으로 증분 탐색법의 한 종류.
* 구간으로 주어진 초기 값에서 근이 포함된 구간을 반으로 나누면서 해를 근사해 나감.

초기값 ($x_l$, $x_u$)

$x_r = \frac{x_l + x_u}{2}$ -> 중간 값

**if** $f(x_l)*f(x_r)$ < 0:

$\quad$ $x_u$ = $x_r$, f($x_u$) = f($x_r$)    #upper limit을 새롭게 정의

**else**  ($f(x_u)*f(x_r)$ < 0):

$\quad$ $x_l$ = $x_r$, f($x_l$) = f($x_r$)    #lower limit을 새롭게 정의

********************************************************************

#### 가위치법(선형 보간법)

* $f(x)$ 의 부호가 바뀌는 구간을 찾는 방법으로 증분 탐색법의 한 종류
* 구간으로 주어진 초기 값에서 $ \{{x_f,f(x_f)\} 과 \{x_u,f(x_u)}\}
 $ 사이의 선을 연결하고 x 축과 만나는 점을 찾는다 . (1차 방정식의 해를 구하는 방법과 동일)
* 알고리즘: 이분법과 비슷하게 iteration (for 문을 활용하여 근사 오차에 도달할때까지 반복 수행한다.)

$\quad$ $ x_r = \frac{f(x_u)x_l - f(x_l)x_u}{f(x_u)-f(x_l)} $

### 개방법

#### 고정점 반복법 

* 순환식을 이용하여 해를 구하는 방법
* 원리 : $ f(x) =  0 $를 $ x = g(x) $와 같은 형태로 변형하여 방정식의 근을 구함. 만약 $ f(x) $를 $ x = g(x) $와 같은 형태로 변형하는것이 불가능할 시에는 $ f(x) $의 양변에 x를 더함으로써 식을 변형 해준다.
* $ x = g(x) $ 방정식의 해 a를 $ g(x) $의 고정점으로 정의함

**순환과정** 
1. 초기 $x_0 $ 설정
2. 이를 통해 $ x_{i+1} = g(x_i)를 구하고, 이 과정을 순차적으로 반복.
3. 이를 수렴 조건 도달 시 까지 반복

*******************************************************************

#### Newton(-Raphson) 법 원리

* 초기 가정값이 $ x_i$ 라면 점 $(x_i,f(x_i))$에 접하는 접선을 활용하는 방법.
* 그리고 그 접선이 $x$축과 만나는 점을 $x_{i+1}$를 근에 대한 개선된 추정값으로 사용.
* $f(x)$의 도함수를 알고 있을 경우 $x_{i+1}$와 $x_i$의 관계식은 다음과 같다.

$\quad$ $x_{i+1} = x_i - \frac{f(x_i)}{f'(x_i)} $

**순환과정**
1. 초기 $x_1$ 설정.
2. $(x_1,f(x_1))$에서$f'(x_1)$도함수 값을 도출.
3. 위의 관계식을 활용하여 새로운$x_2$를 정의.$(x_2 = x_1 - \frac{f(x_1)}{f'(x_2)})$
4. 이를 수렴 조건에 도달시 까지 반복.
