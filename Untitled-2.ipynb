### 3.11 원통형 저장고 내부액체의 부피

#### $$ V = [r^2\cos^{-1}(\frac{r - h}{r}) - (r - h)\sqrt{2rh - h^2}]L $$

* V = 저장고 내부의 액체의 부피
* r = 반지름
* L = 길이
* h = 원통 중심선으로 부터 액체의 깊이

```python
%matplotlib inline
def cyltank(r,L,plot_title):    
    """
    원통안 액체의 높이와 액체의 부피 계산후 그래프 그리기

    parameters
    -----------
        r (_int_): 원통안 액체의 높이
        L (_int_): 원통의 높이
        plot_title (_str_): 그래프 이름
    return
    ------------
    Nan
    """ 
    h = np.linspace(0,6,50)     
    V =  (r**2*np.arccos((r - h)/r) - (r -h)*np.sqrt(2*r*h - h**2)) * L
    plt.figure(figsize=(8,6))
    plt.plot(h,V)
    plt.xlabel(r'$h$')
    plt.ylabel(r'$V$')
    plt.title(plot_title)
    plt.show()
cyltank(3,5,'Volume vs.Depth for Horizontal Cylindrical Tank')
```

### 3.25 transpose temperature

$$ \frac{(F - 32)}{1.8} = C $$
$$ (C \times 1.8) + 32 = F $$

* F = 화씨온도
* C = 섭씨온도

```python
def tempconv(tempin):    
    """_summary_

    Args:
        temperature (_type_): _description_
    """
    global tfdv
    global tcsp
    day = np.array([15,45,75,105,135,165,195,225,255,285,315,345])
    tfdv = np.array([54,60,69,77,87,96,102,101,92,78,63,52])
    tcsp = np.array([-27,-40,-53,-56,-57,-57,-59,-59,-59,-50,-38,-27])
    n = len(day)
    tcdv = np.zeros(n)
    tfsp = np.zeros(n)
    
    if tempin in tfdv:
        for i in range(len(day)):        
            C = (tempin[i] - 32)/1.8
            tcdv[i] = C
        fig, (ax_1,ax_2) = plt.subplots(figsize=(8,6), ncols = 2, nrows = 1)
        ax_1.plot(day,tempin)
        ax_2.plot(day,tcdv)
        plt.show()
    elif tempin in tcsp:
        for i in range(len(day)):
            F = (tempin[i] * 1.8) + 32
            tfsp[i] = F
        fig, (ax_1,ax_2) = plt.subplots(figsize=(8,6), ncols = 2, nrows = 1)
        ax_1.plot(day,tempin)
        ax_2.plot(day,tfsp)
        plt.show()    
    else:
        print('Flase')

tempconv(tcsp)


    

```
